# Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the data

```python
data = pd.read_csv('/Users/nychanthrith/data-science-project/data/Clean data/cleaned_data_January.csv')
```

##  Display basic information about the data

```python
print(data.info())
print(data.describe())
```

## Check for missing values

```python
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)
```

## Plotting histograms for numerical columns

```python
data.hist(bins=30, figsize=(15, 10))
plt.show()
```

## Plotting box plots for numerical columns

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.show()
```

## Plotting correlation matrix

```python
# Exclude non-numeric columns
numeric_data = data.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

# Statistical Analysis

## Calculating mean and standard deviation

```python
mean_value = data['pH'].mean()
std_dev_value = data['pH'].std()
print(f"Mean: {mean_value}, Standard Deviation: {std_dev_value}")
```

## Hypothesis Testing

```python
from scipy import stats
t_stat, p_value = stats.ttest_1samp(data['pH'], 0)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
